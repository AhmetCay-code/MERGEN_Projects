```python
import selenium
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
import random
import time
import pandas as pd
import numpy as np
```

```python
service = Service()
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=service, options=options)
```

```python
driver.get('https://www.mackolik.com/puan-durumu/almanya-bundesliga/fikstur/6by3h89i2eykc341oz7lv1ddd')
time.sleep(5)
```

```python
click_back_week = driver.find_element('xpath','/html/body/div[4]/div[2]/main/div/div[2]/div/div/div[1]')
url_list=[]

for i in range(5):
    click_back_week.click()
    time.sleep(5)
    urls=driver.find_elements('xpath',"//*[@class='p0c-competition-match-list__match-content']")
    for u in range(len(urls)):
        url_list.append(urls[u].find_element('css selector','a').get_attribute('href'))
        
len_url=len(url_list)
```

```python
df = pd.DataFrame(np.zeros((45,23)),columns=['HomeTeam', 'AwayTeam', 'FTHG', 'FTAG','HTHG','HTAG','HS','AS','HST','AST','HF','AF','HC','AC','HY','AY','HR','AR','AvgH','AvgD','AvgA','Avg>2.5','Avg<2.5'])
```

```python
for i in range(len_url):    
    driver.get(url_list[i])
    time.sleep(5)
    
    go_stat = driver.find_element('xpath','/html/body/div[4]/div[1]/div[2]/div/a[4]')
    go_stat.click()
    time.sleep(5)
    
    hteam =driver.find_element('xpath','/html/body/div[4]/div[1]/div[1]/div/div[1]/a[1]').text
    df.HomeTeam[i]=hteam

    ateam =driver.find_element('xpath','/html/body/div[4]/div[1]/div[1]/div/div[1]/a[2]').text
    df.AwayTeam[i]=ateam
    
    fthg =driver.find_element('xpath','/html/body/div[4]/div[1]/div[1]/div/div[2]/div[2]/div[1]/span[1]').text
    df.FTHG[i]=int(fthg)
    
    ftag =driver.find_element('xpath','/html/body/div[4]/div[1]/div[1]/div/div[2]/div[2]/div[1]/span[2]').text
    df.FTAG[i]=int(ftag)
    
    ht =driver.find_element('xpath','/html/body/div[4]/div[1]/div[1]/div/div[2]/div[2]/div[2]').text
    df.HTHG[i]=int(ht[4:5])
    df.HTAG[i]=int(ht[8:9])
    
    general=driver.find_element('xpath','//*[@id="widget-match-live-stats-1"]/div/div/div/div/ul/li[1]/a')
    general.click()
    time.sleep(2)
    
    hcorner =driver.find_element('xpath','//*[@id="widget-match-live-stats-1"]/div/div/div/ul/li[1]/div/table/tbody/tr[12]/td[1]').text
    df.HC[i]=int(hcorner)
    
    acorner =driver.find_element('xpath','//*[@id="widget-match-live-stats-1"]/div/div/div/ul/li[1]/div/table/tbody/tr[12]/td[3]').text
    df.AC[i]=int(acorner)
    
    attack=driver.find_element('xpath','//*[@id="widget-match-live-stats-1"]/div/div/div/div/ul/li[3]')
    attack.click()
    time.sleep(2)
    
    hshot =driver.find_element('xpath','//*[@id="widget-match-live-stats-1"]/div/div/div/ul/li[3]/div/table/tbody/tr[4]/td[1]').text
    df.HS[i]=int(hshot)
    
    ashot =driver.find_element('xpath','//*[@id="widget-match-live-stats-1"]/div/div/div/ul/li[3]/div/table/tbody/tr[4]/td[3]').text
    df.AS[i]=int(ashot)
    
    hshottarget =driver.find_element('xpath','//*[@id="widget-match-live-stats-1"]/div/div/div/ul/li[3]/div/table/tbody/tr[6]/td[1]').text
    df.HST[i]=int(hshottarget)
    
    ashottarget =driver.find_element('xpath','//*[@id="widget-match-live-stats-1"]/div/div/div/ul/li[3]/div/table/tbody/tr[6]/td[3]').text
    df.AST[i]=int(ashottarget)
    
    foul=driver.find_element('xpath','//*[@id="widget-match-live-stats-1"]/div/div/div/div/ul/li[5]')
    foul.click()
    time.sleep(2)
    
    hfoul = driver.find_element('xpath','//*[@id="widget-match-live-stats-1"]/div/div/div/ul/li[5]/div/table/tbody/tr[2]/td[1]').text
    df.HF[i]=int(hfoul)
    
    afoul = driver.find_element('xpath','//*[@id="widget-match-live-stats-1"]/div/div/div/ul/li[5]/div/table/tbody/tr[2]/td[3]').text
    df.AF[i]=int(afoul)
    
    hyellow = driver.find_element('xpath','//*[@id="widget-match-live-stats-1"]/div/div/div/ul/li[5]/div/table/tbody/tr[4]/td[1]').text
    df.HY[i]=int(hyellow)
    
    hred = driver.find_element('xpath','//*[@id="widget-match-live-stats-1"]/div/div/div/ul/li[5]/div/table/tbody/tr[6]/td[1]').text
    df.HR[i]=int(hred)
    
    ayellow = driver.find_element('xpath','//*[@id="widget-match-live-stats-1"]/div/div/div/ul/li[5]/div/table/tbody/tr[4]/td[3]').text
    df.AY[i]=int(ayellow)
    
    ared = driver.find_element('xpath','//*[@id="widget-match-live-stats-1"]/div/div/div/ul/li[5]/div/table/tbody/tr[6]/td[3]').text
    df.AR[i]=int(ared)
    print(i)
    
```

```python
for i in range(len_url):    
    driver.get(url_list[i])
    time.sleep(5)
    
    go_odds = driver.find_element('xpath','/html/body/div[4]/div[1]/div[2]/div/a[3]')
    go_odds.click()
    time.sleep(5)
    
    ### home ###
    #/html/body/div[4]/div[2]/main/div[1]/div[2]/div/div[2]/div/div[1]/div/ul/li[1]/div/div/ul[1]/li[1]/span[2]
    boolean1=len(driver.find_elements('xpath','/html/body/div[4]/div[2]/main/div/div[2]/div/div[2]/div/div[1]/div/ul/li[1]/div/div/ul/li[1]/a/span[2]'))>0

    if boolean1 == False:
        h = driver.find_element('xpath','/html/body/div[4]/div[2]/main/div/div[2]/div/div[2]/div/div[1]/div/ul/li[1]/div/div/ul/li[1]/span[2]').text
    else:
        h=driver.find_element('xpath','/html/body/div[4]/div[2]/main/div/div[2]/div/div[2]/div/div[1]/div/ul/li[1]/div/div/ul/li[1]/a/span[2]').text

    if h=='-':
        h=1
    else:
        h=h
    
    df.AvgH[i]=h
    
    ### draw ###
    #/html/body/div[4]/div[2]/main/div[1]/div[2]/div/div[2]/div/div[1]/div/ul/li[1]/div/div/ul[1]/li[2]/span[2]
    boolean2=len(driver.find_elements('xpath','/html/body/div[4]/div[2]/main/div/div[2]/div/div[2]/div/div[1]/div/ul/li[1]/div/div/ul[1]/li[2]/a/span[2]'))>0
    
    if boolean2 == False:
        d =driver.find_element('xpath','/html/body/div[4]/div[2]/main/div/div[2]/div/div[2]/div/div[1]/div/ul/li[1]/div/div/ul[1]/li[2]/span[2]').text
    else:
        d =driver.find_element('xpath','/html/body/div[4]/div[2]/main/div/div[2]/div/div[2]/div/div[1]/div/ul/li[1]/div/div/ul[1]/li[2]/a/span[2]').text

    if d=='-':
        d=1
    else:
        d=d
    
    df.AvgD[i]=d
    
    ### away ###
    #/html/body/div[4]/div[2]/main/div[1]/div[2]/div/div[2]/div/div[1]/div/ul/li[1]/div/div/ul[1]/li[3]/span[2]
    boolean3=len(driver.find_elements('xpath','/html/body/div[4]/div[2]/main/div/div[2]/div/div[2]/div/div[1]/div/ul/li[1]/div/div/ul[1]/li[3]/a/span[2]'))>0

    if boolean3 == False:
        a = driver.find_element('xpath','/html/body/div[4]/div[2]/main/div/div[2]/div/div[2]/div/div[1]/div/ul/li[1]/div/div/ul[1]/li[3]/span[2]').text
    else:
        a=driver.find_element('xpath','/html/body/div[4]/div[2]/main/div/div[2]/div/div[2]/div/div[1]/div/ul/li[1]/div/div/ul[1]/li[3]/a/span[2]').text

    if a=='-':
        a=1
    else:
        a=a
    
    df.AvgA[i]=a
    
    ### under-over ###
    
    ou =driver.find_elements('xpath','//*[@class="widget-iddaa-markets__market-item"]')
    ou_list=[]
    for m in range(len(ou)):
        ou_list.append(ou[m].text)
    for k in range(len(ou_list)):
        if ou_list[k][0:11]=='2,5 ALT/ÃœST':
            u=ou_list[k][28:32]
            o=ou_list[k][37:41]
    if o=='-':
        o=1
    else:
        o=o
    
    if u=='-':
        u=1
    else:
        u=u
    
    df['Avg>2.5'][i]=o
    df['Avg<2.5'][i]=u
    print(i)
```

```python
df= df.iloc[::-1]
df.reset_index(drop=True)
```

```python
df.to_csv('Test_Set/germany_last_four_weeks.csv',index=False)
```
