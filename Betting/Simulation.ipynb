```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
def result_generator(p,quantity_result):
    result=1
    for i in range(quantity_result):
        temp=np.random.choice(np.arange(0, 2), p=[(p/100), (1-(p/100))])
        result=temp*result
    return result
def odd_generator(quantity_odd):
    odd=1
    for i in range(quantity_odd):
        temp=random.uniform(1.67 , 2.41)
        odd=temp*odd
    return odd
```

```python
numofsim=1000
coefficient_of_gain=10
numberof_betting_slip=15
for betnumber in range(1,6):
    suc_ratio=[]
    model_acc=[]
    for model_accuracy in range(5,30):
        suc=0
        fail=0
        for j in range(numofsim):
            capital = 100
            for i in range(numberof_betting_slip):
                bet=capital*0.75
                result=result_generator(model_accuracy,betnumber)
                odd = odd_generator(betnumber)
                if result == 1:
                    capital=(capital-bet)+(bet*odd)
                else:
                    capital=(capital-bet)
                    if capital<2:
                        break        
            if capital>(100*coefficient_of_gain):
                suc=suc+1
            else:
                fail=fail+1
        print('% '+str(round(suc/(suc+fail)*100,2))+' SUCCESS RATIO at '+str(100-model_accuracy)+ ' accuracy of model for '+str(numberof_betting_slip)+' betting slip and each has ' + str(betnumber)+ ' games')    
        print('--------------------------------------------------------')
        suc_ratio.append(round(suc/(suc+fail)*100,2))
        model_acc.append(100-model_accuracy)
    plt.figure(figsize=(20,10),dpi=80)
    plt.plot(model_acc,suc_ratio)
    plt.title('Bet Success Ratio - Model Accuracy Plot according to '+str(numberof_betting_slip)+' betting slip and each has '+ str(betnumber)+ ' games')
    plt.xlabel("% accuracy ")
    plt.ylabel("% succes ratio")
    plt.xlim([71,95])
    plt.ylim([0, 100])
    plt.show()
```
